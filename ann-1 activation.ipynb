```python
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
```

```python
def binaryStep(x):
    ''' It returns '0' is the input is less then zero otherwise it returns one '''
    return np.heaviside(x,1)
```

```python
x = np.linspace(-10, 10)
plt.plot(x, binaryStep(x))
plt.axis('tight')
plt.title('Activation Function :binaryStep')
plt.show()
```

```python
def linear(x):
    ''' y = f(x) It returns the input as it is'''
    return x
```

```python
x = np.linspace(-10, 10)
plt.plot(x, linear(x))
plt.axis('tight')
plt.title('Activation Function :Linear')
plt.show()
```

```python
def sigmoid(x):
    ''' It returns 1/(1+exp(-x)). where the values lies between zero and one '''

    return 1/(1+np.exp(-x))
```

```python
x = np.linspace(-10, 10)
plt.plot(x, sigmoid(x))
plt.axis('tight')
plt.title('Activation Function :Sigmoid')
plt.show()
```

```python
def tanh(x):
    ''' It returns the value (1-exp(-2x))/(1+exp(-2x)) and the value returned will be lies in between -1 to 1.'''

    return np.tanh(x)
```

```python
x = np.linspace(-10, 10)
plt.plot(x, tanh(x))
plt.axis('tight')
plt.title('Activation Function :Tanh')
plt.show()
```

```python
def RELU(x):
    ''' It returns zero if the input is less than zero otherwise it returns the given input. '''
    x1=[]
    for i in x:
        if i<0:
            x1.append(0)
        else:
            x1.append(i)

    return x1
```

```python
x = np.linspace(-10, 10)
plt.plot(x, RELU(x))
plt.axis('tight')
plt.title('Activation Function :RELU')
plt.show()
```

```python
def leaky_relu(x, alpha=0.05):
    return np.maximum(alpha * x, x)
```

```python
x = np.linspace(-10, 10, 100)
y = leaky_relu(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('leaky_relu(x)')
plt.title('Leaky ReLU Activation Function')
plt.grid(True)
plt.show()
```

```python


def elu(x, alpha=1.0):
    return np.where(x > 0, x, alpha * (np.exp(x) - 1))

def swish(x, beta=1.0):
    return x * (1 / (1 + np.exp(-beta * x)))

# Generate x values
x = np.linspace(-5, 5, 100)

# Calculate y values for ELU and Swish
y_elu = elu(x)
y_swish = swish(x)

# Plot ELU activation function
plt.figure(figsize=(8, 4))
plt.plot(x, y_elu, label='ELU', color='blue')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Exponential Linear Unit (ELU) Activation Function')
plt.legend()
plt.grid(True)
plt.show()

# Plot Swish activation function
plt.figure(figsize=(8, 4))
plt.plot(x, y_swish, label='Swish', color='red')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Swish Activation Function')
plt.legend()
plt.grid(True)
plt.show()
```
